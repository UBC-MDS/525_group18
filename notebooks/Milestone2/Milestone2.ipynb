#### Downloading the Data

````python


# Necessary metadata
article_id = 14226968  # this is the unique identifier of the article on figshare
url = f"https://api.figshare.com/v2/articles/{article_id}"
headers = {"Content-Type": "application/json"}
output_directory = "/srv/data/my_shared_data_folder/"

````

````python
response = requests.request("GET", url, headers=headers)
data = json.loads(response.text)  # this contains all the articles data, feel free to check it out
files = data["files"]             # this is just the data about the files, which is what we want
files
````

````python

files_to_dl = ["combined_model_data_parti.parquet.zip"]  ## Please download the partitioned 
for file in files:
    if file["name"] in files_to_dl:
        os.makedirs(output_directory, exist_ok=True)
        urlretrieve(file["download_url"], output_directory + file["name"])

````

````python


with zipfile.ZipFile(os.path.join(output_directory, "combined_model_data_parti.parquet.zip"), 'r') as f:
    f.extractall(output_directory)

````

### 6. Wrangle the data in preparation for machine learning

````python


%%time
%%memit

# Passing your credentials 
## IMPORTANT: make sure you dont include you secret and key when submitting the notebook
aws_credentials ={"key": "use your own","secret": "use yur own"} ## dont include you secret and key when submitting the notebook
df = pd.read_parquet("s3://mds-s3-student65/combined_model_data_parti.parquet", storage_options=aws_credentials)
````

````python
df.head()
````

````python
df
````

````python
df.shape
````

Now let us proceed with wrangling the rainfall dataframe to only include records for Sydney, Australia.

syd_lat = -33.86

syd_lon = 151.21

````python
syd_lat = -33.86

syd_lon = 151.21
````

````python
sydney = df.query("lat_min < @syd_lat < lat_max & lon_min < @syd_lon < lon_max")
sydney
````

Futhermore, now let us proceed with dropping the columns that are not needed in the analysis, such as latitude and longtitude values for the city Sydney.

````python
sydney_df = sydney.drop(columns = ["lat_min", "lat_max", "lon_min", "lon_max"])
sydney_df
````

Now, let's proceed with setting the time as the index of the dataframe.

````python
sydney_index = sydney_df.set_index("time")
sydney_index
````

It is it time for us to tranform the data so that is appropriate for training a machine learning model. Hence we are going to pivot the column `model` and assign the values of `rain (mm/day)` for each model in study.

````python
sydney_pivot = sydney_index.pivot(columns='model', values = "rain (mm/day)")
sydney_pivot
````

After pivoting the data we see that are missing values, hence let us resample the data and aggragate it with the `mean` function.

````python
df_sydney = sydney_pivot.resample("1D").mean()
df_sydney
````

Now we need to add an extra column for the target, which is saved in the s3 bucket. The file is called `observed_daily_rainfall_SYD.csv`, let us load the for this.

````python
observed_df = pd.read_csv("s3://mds-s3-student65/observed_daily_rainfall_SYD.csv", storage_options=aws_credentials)
````

````python
observed_df
````

````python
observed_df.shape
````

Let us add this the observed rainfall data as a column in our wrangled Sydney dat set of rainfall values for different models.

````python
df_sydney["observed"] = observed_df[["rain (mm/day)"]].values
````

````python
df_sydney
````

````python
df_sydney
````

Finally our dataset is ready for training model of rainfall prediction with machine learning.

````python
df_sydney.to_csv("s3://mds-s3-student65/output/ml_data_SYD.csv", storage_options=aws_credentials)
````
